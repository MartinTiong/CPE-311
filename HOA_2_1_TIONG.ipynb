```python
A = [10, 8, 5, 3]
B = []
C = []

def Movement(Source, Desti, src, dst ):
 disk = Source.pop()
 Desti.append(disk)

 print(f"A: {A}, B: {B} C: {C} " )

def hanoi(n, Source, Aux, Desti, src, ax, dst):
    if n == 1:
        Movement(Source, Desti, src, dst)
    else:
        hanoi(n-1, Source, Desti, Aux, src, dst, ax)
        Movement(Source, Desti, src, dst)
        hanoi(n-1, Aux, Source, Desti, ax, src, dst)

hanoi(len(A), A, B, C, "A", "B", "C")


```

This code integrates the usage of the pop and append command to "pop" the value (disk) currently on top of the source peg. After it is removed, append is used to transport said disk on to the destination peg. The hanoi function contains the main logic and follows the tower of hanoi algorithm. It has a if-else loop to continously check the disks whether to move them or not (n). The function call uses len command as a way to signify the number of disks, for the n to know when to stop continuing.
