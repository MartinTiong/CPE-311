```python
import pandas as pd

df = pd.read_csv("/content/Meteorite_Landings.csv")
```

##limiting the years

```python
df["year"] = df["year"].str.slice(0, 4)
df["year"] = pd.to_numeric(df["year"], errors="coerce")

df_2005_2009 = df[(df["year"] >= 2005) & (df["year"] <= 2009)]
```

```python
pivot = pd.pivot_table(
    df_2005_2009,
    index="year",
    columns="fall",
    values="mass (g)",
    aggfunc=["count", lambda x: x.quantile(0.95)]
)

pivot
```

##Exercise 5

```python
taxi = pd.read_csv("/content/2019_Yellow_Taxi_Trip_Data.csv")
```

```python
taxi["tpep_dropoff_datetime"] = pd.to_datetime(taxi["tpep_dropoff_datetime"])
taxi = taxi.set_index("tpep_dropoff_datetime")


```

```python
hourly = taxi.resample("H").agg({
    "trip_distance": "sum",
    "fare_amount": "sum",
    "tolls_amount": "sum",
    "tip_amount": "sum"
})

hourly.head()
```

```python
hourly = taxi.resample("H").agg({
    "trip_distance": "sum",
    "fare_amount": "sum",
    "tolls_amount": "sum",
    "tip_amount": "sum"
})

hourly.head()
```

```python
top_tip_hours = hourly.nlargest(5, "tip_amount")
top_tip_hours
```
